### Exercise 5.1
The requested sampling of probability density function for the atomic orbitals is performed with the code in the file **ex_5.cpp**.<br>
The code is done in such a way that it is easy to add a new pdf to sample. To add a pdf one should provide a function of type `pdf_type` with prototype `double pdf(R3_point)` and add a new part in function `get_orbital_parameters`. The last function is a function that is called before performing Metropolis to get the address of the pdf implementation and a Metropolis step suitable for that pdf.<br>
The sampled points are saved in files **./out/ \*\_orbital\_sampling.dat**. The averages are shown in files **./out/ \*\_averages.dat**. As usual averages are performed with a fixed number L of data per block and plotted for a varying number of blocks. <br>
It has been chosen $L=50$ and $M=2\cdot10^5$ as a maximum number of points sampled. Sparse averaging is performed using one point every 10.

The following code plot the values for $\left\langle r \right\rangle_{\Psi}$ as a function of N for some orbitals. The sampling has been performed for orbitals 1s, 2s, 2p, 3d and 3p; all of them are in $j_z = 0$ state. The exact value for $\left\langle r \right\rangle_{\Psi}$ is plotted for each orbital.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
orbital_list = ["1s","2s","2p","3d", "3p"]
exact_values_list = [1.5,6,5.0,10.5,12.5]
plt.figure(figsize=(15, 18))
folder = "./out/"
count= 1

for orbital in orbital_list:
    data = np.loadtxt(folder+orbital+"_averages.dat", dtype=float)
    plt.subplot(3,2,count)
    
    plt.title("Sampling of the orbital "+orbital)
    plt.xlabel("# blocks")
    plt.ylabel('<r>')
    plt.xscale("log")
    plt.errorbar(data[:,0],data[:,1], xerr=None, yerr=data[:,2])
    plt.plot([0,data[-1,0]],[exact_values_list[count-1],exact_values_list[count-1]])
    count = count +1

plt.show()
```

Some observations:
 - The "step size" of a proposed move strongly depends on the kind of pdf sampled because every pdf has a different spatial scale. This means that a different steps l must be set for every pdf: this is achieved in the code by the function `get_orbital_parameters`.
 - The number L of measure in each block should be large enough that measures in different blocks are uncorrelated. This means that L should be larger than the auto-correlation time constant $t_c$. We haven't computed that here but (also comparing with $t_c$ for MD code in ex. 7) $L = 50$ seems like a good choiche. Furthermore it gives meaningful results.
 - Choosing a starting point far from the origin doesn't seem to change much the performances of the algorithm. It has been tried to set $P_0 = (100,-100,200)$ and the results are still realiable when using a large number of blocks (i.e. sampled points which mean more Montecarlo steps). For a few number of points results are worse than those obtained choosing a fair starting point: the algorithm takes more to get closer to the zone of high probability and thus the first sampled points are highly unrealiable. This shortcoming can be cured by choosing a higher equilibration time.
 - A gaussian Metropolis step has been tried by proposing a random step sampled from $\mathcal{N}(0,\frac{l}{1.7})$ (l being the step of the uniform proposed move and 1.7 a constant factor to keep acceptance under control). As might be expected there are no differences from the previous case: the normal distribution still satisfies $T(x,y)=T(y,x) \propto exp(-\frac{(x-y)^2}{2\cdot\sigma})$ and thus is a valid transition probability.

It is also possible to plot the 3D shape of the orbitals using some sampled points.

```python
N = 1000 #points to use for the plot

for orbital in orbital_list:
    data = np.loadtxt(folder+orbital+"_orbital_sampling.dat", dtype=float)[0:N,:]
    
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(data[:,0], data[:,1], data[:,2], c=data[:,2], marker='.',label = orbital)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.title("Probability for orbital "+orbital)
    ax.legend();
    ax.view_init(10, 30)
    
plt.show()
```
