### Exercise 5.1
The requested sampling of probability density function for the atomic orbitals is performed with the code in the file **ex_5.cpp**.<br>
The code is done in such a way that it is easy to add a new pdf to sample. To add a pdf one should provide a function with prototype *double pdf(R3_point)* and add a new part in function get_orbital_parameter. The last function is a function that is called before performing Metropolis to get the address of the pdf implementation and a Metropolis step suitable for that pdf.<br>
The sampled points are saved in files \*\_orbital\_sampling.dat. The averages are shown in files \*\_averages.dat. As usual averages are performed with a fixed number L of data per block and plotted for a varying number of blocks. <br>
It has been chosen $L=50$ and $M=10^4$ as a maximum number of points sampled. Sparse averaging is performed using one point every 10.

The following code plot the values for $\left\langle r \right\rangle_{\Psi}$ as a function of N for some orbitals. The sampling has been performed for orbitals 1s, 2s, 2p, 3d and 3p; all of them are in $j_z = 0$ state.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
orbital_list = ["1s","2s","2p","3d", "3p"]
exact_values_list = [1.5,6,5.0,10.5,12.5]
plt.figure(figsize=(15, 15))
count= 1

for orbital in orbital_list:
    data = np.loadtxt(orbital+"_averages.dat", dtype=float)
    plt.subplot(3,2,count)
    
    plt.title("Sampling of the orbital "+orbital)
    plt.xlabel("# blocks")
    plt.ylabel('<r>')
    plt.xscale("log")
    plt.errorbar(data[:,0],data[:,1], xerr=None, yerr=data[:,2])
    plt.plot([0,10000],[exact_values_list[count-1],exact_values_list[count-1]])
    count = count +1

plt.show()
```

It is also possible to plot the 3D shape of the orbitals using some sampled points.

```python
from mpl_toolkits.mplot3d import Axes3D
N = 1000 #points to use for the plot

for orbital in orbital_list:
    data = np.loadtxt(orbital+"_orbital_sampling.dat", dtype=float)[0:N,:]
    
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(data[:,0], data[:,1], data[:,2], c=data[:,2], marker='.',label = orbital)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.title("Probability for orbital "+orbital)
    ax.legend();
    ax.view_init(10, 30)
    
plt.show()
```
