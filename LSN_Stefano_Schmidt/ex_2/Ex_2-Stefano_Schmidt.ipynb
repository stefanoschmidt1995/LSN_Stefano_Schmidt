### Exercise 2.1
The integral $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$ is computed as required in two ways:
- Uniform sampling: points $x_i$ are sampled uniformly in [0,1] and the value of the integral is obtained averaging the quantities $y^{UN}_i = \frac{\pi}{2}\cos(\pi x_i/2)$
- Importance sampling: points are sampled in [0,1] following the pdf $p(x) = \frac{2}{3} \cdot (1-x^2)$. The pdf has been chosen because its shape is "similar" to the integrand function. The integral is then obtained averaging the quantities $y^{IS}_i = \frac{\frac{\pi}{2}\cos(\pi x_i/2)}{p(x)} $.

The performances of the two methods are compared by averaging the data with a fixed number of data per blocks ($L=50$) and varying the number N of blocks. In this way the performances are evaluated with different number of samples $M = N*L$.<br>
The programs produces the output in the two files **unif_sampling.dat** and **imp_sampling.dat** holding the average values of the integral (with uncertainties) as a function of N (number of blocks). The following code produces the required plots.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
unif_samp_data = np.loadtxt("unif_sampling.dat")
imp_samp_data = np.loadtxt("imp_sampling.dat")

plt.figure(figsize=(15, 4))

plt.subplot(1,2,1)
plt.errorbar(unif_samp_data[:,0],unif_samp_data[:,1], yerr=unif_samp_data[:,2])
plt.plot([0,imp_samp_data[-1,0]],[1,1])
plt.xscale('log')
plt.title("Uniform sampling")
plt.xlabel("# blocks")
plt.ylabel("Integral value")

plt.subplot(1,2,2)
plt.errorbar(imp_samp_data[:,0],imp_samp_data[:,1], yerr=imp_samp_data[:,2])
plt.plot([0,imp_samp_data[-1,0]],[1,1])
plt.xscale('log')
plt.title("Importance sampling")
plt.xlabel("# blocks")
plt.ylabel("Integral value")

plt.show()
```

It is worth to be noticed that the importance sampling method produces a realiable output (i.e. compatible with the actual value within the uncertainties) even using a few data ($N\simeq10$); in order to have a good measure of the integral with uniform sampling a bigger amount of data ($N\simeq150$) is required.

### Exercise 2.2
The two versions or Random Walk are simulated $N_{sim} = 2000$ times. Each simulation consists in $n_{step} = 1000$ time steps. <br>
For each time step $i$ is performed the average of the quantity $\bar{r}^2 = \langle |\vec{r}_i|^2 \rangle_{N_{sim}}$ over all the simulation performed.The uncertainties are computed using data blocking with $N_{blocks} = 100 $ blocks. <br>
In file **r_average.dat** are printed as a function of time step i the values of $\sqrt{\bar{r_i}^2}$ for the two versions of Random Walk along with their errors.

```python
RW_data = np.loadtxt("r_average.dat", skiprows=1)

plt.figure(figsize=(15, 7))

plt.subplot(1,2,1)
plt.errorbar(RW_data[:,0],RW_data[:,1], yerr=RW_data[:,2])
start_t = 10
plt.plot(RW_data[:,0], np.sqrt(RW_data[:,0]), linewidth=4)
plt.plot(RW_data[start_t:-1,0], .59*np.sqrt(RW_data[start_t:-1,0])*np.sqrt(2*np.log(np.log(RW_data[start_t:-1,0]))), linewidth=2, c= "r")
plt.plot(RW_data[start_t:-1,0], 2*np.sqrt(RW_data[start_t:-1,0])-0.59*np.sqrt(RW_data[start_t:-1,0])*np.sqrt(2*np.log(np.log(RW_data[start_t:-1,0]))), linewidth=2, c="r")

plt.title("Discrete Random Walk")
plt.xlabel("time step")
plt.ylabel("$(<{r}^2>)^{1/2}$")

plt.subplot(1,2,2)
plt.errorbar(RW_data[:,0],RW_data[:,3], yerr=RW_data[:,4])
plt.plot(RW_data[:,0], np.sqrt(RW_data[:,0]), linewidth=4)
plt.plot(RW_data[start_t:-1,0], .59*np.sqrt(RW_data[start_t:-1,0])*np.sqrt(2*np.log(np.log(RW_data[start_t:-1,0]))), linewidth=2, c= "r")
plt.plot(RW_data[start_t:-1,0], 2*np.sqrt(RW_data[start_t:-1,0])-0.59*np.sqrt(RW_data[start_t:-1,0])*np.sqrt(2*np.log(np.log(RW_data[start_t:-1,0]))), linewidth=2, c="r")
plt.title("Continuos Random Walk")
plt.xlabel("time step")
plt.ylabel("$(<{r}^2>)^{1/2}$")

plt.show()
```

It has been drawn (orange) also the curve $\sqrt{\bar{r_i}^2} (i) = \sqrt{i}$. As can be seen both experimental curves are well fitted by that. This clearly indicates the typical RW scaling of average displacement. <br>
It is worth to be noticed that for a Wiener process $W_t$ the following relations holds: <br> <center>$\limsup\limits_{x\rightarrow \infty} \frac{W_t}{\sqrt{2\cdot t \cdot \mathrm{log log} t}} = 1$ 
and $\liminf\limits_{x\rightarrow \infty} \frac{W_t}{\sqrt{2\cdot t \cdot \mathrm{log log} t}} = -1$
</center>
This gives the asymptotic law for the maximum displacement from the average of $\sqrt{\bar{r_i}^2}$. This is of course also a scaling law for the statistical uncertainties $\sigma_{|r_i|}$ of $\sqrt{\bar{r_i}^2}$: $\sqrt{\bar{r_i}^2} + \sigma_{|r_i|} \backsim c \cdot \sqrt{2\cdot t \cdot \mathrm{log log} t}$ where $c$ is a constant $\leq 1$ (here $c = 0.59$). The two scaling for the upper and lower uncertainties are plotted in red in the two plots above.

It can be useful to visualize a sample of a trajectory of a discrete RW. This feature is included in the code that saves in file **continuos_RW.dat** at every time the position of the walker. The 3D plot is made below and a red marker is put on the origin of the walk.

```python
from mpl_toolkits.mplot3d import Axes3D

RW_data = np.loadtxt("continuos_RW.dat", dtype=float)

ax = Axes3D(plt.figure(figsize=(8, 8)))
ax.scatter(RW_data[:,0], RW_data[:,1], RW_data[:,2], c=RW_data[:,2], marker='.')
ax.scatter([0], [0], [0], c='red', marker='o', s =40)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```
