### Exercise 11.1
The required linear regression is performed for different choices of parameters $N_{\mathrm{train}}$, $N_{\mathrm{epochs}}$ and noise sigma. Some "experiments" are done below.

```python
#really bad case: few noisy data and few training epochs
N_train = 10
sigma = 2.0
N_epochs = 2

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true, x_train, y_train)
```

```python
#slightly better: more training epochs
N_train = 10
sigma = 2.0
N_epochs = 20

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true, x_train, y_train)
```

From the two plots above we learnt about underfitting: too few training is done and thu the model's parameters are not optimized. Regardless whether the model might be good or not for the given problem, underfitting means that the model is "not doing his best".

```python
#almost the best possible: much more training epochs
N_train = 10
sigma = 2.0
N_epochs = 200

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true, x_train, y_train)
```

```python
#overfitting?
N_train = 10
sigma = 2.0
N_epochs = 2000

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true, x_train, y_train)
```

Comparing the last plot with the previous one we see that overfitting is an actual problem in machine learning: we passed from a good agreement with $N_{\mathrm{epochs}} = 200$ to a bad agreement with $N_{\mathrm{epochs}} = 2000$. The model has learnt the data, not the underlying pattern.

```python
#having a lot of data can help do less training?
N_train = 1000
sigma = 2.0
N_epochs = 2

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true)
```

Indeed having a lot of data can help! The result is (almost) the same as the one with $N_{\mathrm{epochs}} = 200$ and $N_{\mathrm{train}} = 10$ but here there have been only two epochs. This is not weird: an epoch swipes through all data; having many data makes an epoch more effective.

```python
#data very very noisy: is ML possible?
N_train = 100
sigma = 50.0
N_epochs = 200

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true, x_train[0:20], y_train[0:20])
```

Probably here there is very few to learn: the actual function is washed out by noise and it's impossible to recognize any pattern. The predicted function is almost meaningless.

```python
#perfect case: no noise
N_train = 100
sigma = 0.0
N_epochs = 1000

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train,10, sigma)
model = fit_NN_model(x_train, y_train, N_epochs)[0]
plot_predictions(model, x_valid, y_true, x_train[0:20], y_train[0:20])
```

The model is able to fully reproduce the data. The loss function is zero and there is no concern of overfitting.

### Exercise 11.3
The sinuoisodal function is defined below. The code in NN_helper.py is done in such a way that it's easy to generalize to a higher dimensions features space.

```python
N_train = 400
sigma = 0.5
```

```python
#defining sinusoidal function
f_sin = lambda x: np.sin(x[0]**2 + x[1]**2)
```

```python
#usual three layers model
N_epochs = 100

model= keras.Sequential()
model.add(Dense(10, input_shape=(2,)))
model.add(Dense(10, activation="sigmoid"))
model.add(Dense(1,activation="relu"))

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train, 100, sigma, function=f_sin, ndim=2, x_range=1.5)
model = fit_NN_model(x_train, y_train, N_epochs, model)[0]
plot_3D_predictions(model, x_valid, y_true)
```

The model doesn't seem to generalize well for two dimensional functions... Probably it's too few general: it should be worth to try with a different more flexible model.

```python
#more complicated five layers model
N_epochs = 1000

model= keras.Sequential()
model.add(Dense(10, input_shape=(2,)))
model.add(Dense(20, activation="relu"))
model.add(Dense(20, activation="sigmoid"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1))

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train, 100, sigma, function=f_sin, ndim=2, x_range=1.5)
model = fit_NN_model(x_train, y_train, N_epochs, model)[0]
loss = model.evaluate(x_valid, y_true, batch_size=32, verbose=0)[0]
plot_3D_predictions(model, x_valid, y_true)
```

```python
#very complicated seven layers model
N_epochs = 2000

model= keras.Sequential()
model.add(Dense(10, input_shape=(2,)))
model.add(Dense(20, activation="relu"))
model.add(Dense(20, activation="sigmoid"))
model.add(Dense(10, activation="relu"))
model.add(Dense(20, activation="sigmoid"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1))

x_train, y_train, x_valid, y_valid, y_true = create_training_set(N_train, 100, sigma, function=f_sin, ndim=2, x_range=1.5)
model = fit_NN_model(x_train, y_train, N_epochs, model)[0]
loss = model.evaluate(x_valid, y_true, batch_size=32, verbose=0)[0]
plot_3D_predictions(model, x_valid, y_true)
```

The model here is much harder to fit. To obtain some good results it has been necessary to use a complicated model with a huge number of parameters and a long traininig procedure. Again it is worth to be noticed that increasing the complexity of model doesn't necessarly increases the performances: the seven layers model has been fitted with 2000 epochs and despite this it gives a loss more than the one of the five layers model (trained with "only" 1000 epochs)
