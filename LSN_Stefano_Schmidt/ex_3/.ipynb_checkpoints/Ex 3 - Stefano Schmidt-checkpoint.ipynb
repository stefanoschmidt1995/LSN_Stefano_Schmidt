### Exercise 3.1
The pricing of the options is done as required and an output file **price.dat** is produced by the code with the following header: <br>
**N | C direct | sigma C direct | C samples | sigma C sample | P direct | sigma P direct | P sample | sigma P sample** <br>
Here C stands for Call Option price and P fo Put Option price; direct means that the value is obtained sampling GBM at final time and sample means that the time interval has been divided in 100 steps and sampled stepwise. <br>
The averages are done over M samples of the GBM with an increasing number N of blocks (each block holding a fixed number $L = 50$) so that $M = N \cdot L$. In the following code the different plots are performed and compared with the exact value given by the formula of Black and Scholes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
price_data = np.loadtxt("prices.dat", skiprows=1)
plt.figure(figsize=(15, 15))

plt.subplot(2,2,1) #C direct
plt.title("Call price - direct sampling")
plt.xlabel("# blocks")
plt.ylabel('Call Price')
plt.xscale("log")
plt.errorbar(price_data[:,0],price_data[:,1], xerr=None, yerr=price_data[:,2])
plt.plot([0,10000],[14.97579,14.97579])

plt.subplot(2,2,2) #C sample
plt.title("Call price - stepwise sampling")
plt.xlabel("# blocks")
plt.ylabel('Call Price')
plt.xscale("log")
plt.errorbar(price_data[:,0],price_data[:,3], xerr=None, yerr=price_data[:,4])
plt.plot([0,10000],[14.97579,14.97579])

plt.subplot(2,2,3) #C direct
plt.title("Put price - direct sampling")
plt.xlabel("# blocks")
plt.ylabel('Put Price')
plt.xscale("log")
plt.errorbar(price_data[:,0],price_data[:,5], xerr=None, yerr=price_data[:,6])
plt.plot([0,10000],[5.45953,5.45953])

plt.subplot(2,2,4) #C sample
plt.title("Call price - stepwise sampling")
plt.xlabel("# blocks")
plt.ylabel('Put Price')
plt.xscale("log")
plt.errorbar(price_data[:,0],price_data[:,7], xerr=None, yerr=price_data[:,8])
plt.plot([0,10000],[5.45953,5.45953])

plt.show()
```
