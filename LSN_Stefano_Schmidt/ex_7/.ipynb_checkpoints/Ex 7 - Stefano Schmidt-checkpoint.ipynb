### Exercise 7.1
The code in **ex_7.cpp** saves the istantaneous values of energy and pressure to files **e.out** and **p.out**. Those files are read in input to to compute the autocorelation function by the function get_autocorrelation() below.<br>
The computations are done with $N=2 \cdot 10^5$ points.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_autocorrelation(input_files, times):
    """This functions takes data from input_file and computes the autocorrelation function
    for the times specified in times.
        Input:  input_file (string): keeps filename of the source of the data
                times (n,): keeps the times at which compute autocorrelation function
        Output: autocorr (n,): the value of autocorrelation function for the times specified in times
    """
    data = np.loadtxt(input_files)
    data_avg = np.average(data)
    data_var = np.var(data)
    autocorr = np.zeros(times.shape)
    
    for t in range(np.size(times)):
        cross_term = 0
        avg_t = 0; avg_ret_t =0;
        index = 0
        while t+index < np.size(data):
            cross_term = cross_term + (data[index]*data[t+index])
            avg_t = avg_t + data[index]
            avg_ret_t = avg_ret_t + data[t+index]
            index = index+1
        if index ==0:
            break

        cross_term = cross_term /float(index)
        avg_t = avg_t /float(index)
        avg_ret_t = avg_ret_t /float(index)
        autocorr[t] = (cross_term - avg_t*avg_ret_t )/data_var
    return autocorr
```

```python
def get_correlation_time(times, autocorr):
    """Returns the correlation time for data as a function of time
    Input:  times (n,): keeps the times at which the autocorrelation function is computed
            autocorr (n,): keeps the value of autocorrelation function
    Output: t_corr (): correlation time (i.e. time s.t data = exp(-times/t_corr))
    """
    from scipy.optimize import curve_fit
    exp_fun = lambda x, t_c: np.exp(-x/t_c)
    popt, pcov = curve_fit(exp_fun, times, autocorr, p0=(1))
    return popt[0]
        
    
```

```python
#computing autocorrelations functions
t_max =  500
times = np.arange(0,t_max, step = 1)
e_autocorr = get_autocorrelation("e.out",times)
p_autocorr = get_autocorrelation("p.out",times)
```

```python
#plotting and fitting autocorrelation functions
t_max =  500
times = np.arange(0,t_max, step = 1)
t_max_fit = 250

plt.figure(figsize=(15, 6))

#energy
plt.subplot(1,2,1)
plt.plot(times, e_autocorr, label="Computed")
e_t_corr = get_correlation_time(times[0:t_max_fit], e_autocorr[0:t_max_fit])
plt.plot(times, np.exp(-times/e_t_corr), label ="Fitted (t_c ="+str(np.round(e_t_corr,3))+")")
plt.xlabel("time")
plt.ylabel("Autocorellation function")
plt.legend()
plt.title("Autocorrelation of energy")

#pressure
plt.subplot(1,2,2)
plt.plot(times, p_autocorr, label="Computed")
p_t_corr = get_correlation_time(times[0:t_max_fit], p_autocorr[0:t_max_fit])
plt.plot(times, np.exp(-times/p_t_corr), label ="Fitted (t_c ="+str(np.round(p_t_corr,3))+")")
plt.xlabel("time")
plt.ylabel("Autocorellation function")
plt.legend()
plt.title("Autocorrelation of pressure")

plt.show()
```

The exponential fit works well for early times but fails to reproduce the decay of autocorrelation at intermediate times. The correlation time found in both cases is around 20. This suggest that using $L \simeq 50$ measures per block provides quite uncorrelated values for the averages of different blocks.

The code produces also the file **avg_e.out** and **avg_p.out** which keeps the uncertainties of average quantities (energy and pressure) as a function of L (# of data per block). The number of points sampled is $N=2 \cdot 10^5$. The plots are done below.

```python
e_errors = np.loadtxt("avg_e.out", dtype=float)
p_errors = np.loadtxt("avg_p.out", dtype=float)

#print(e_errors)

plt.figure(figsize=(15, 6))

#energy errors
plt.subplot(1,2,1)
plt.title("Energy - errors in averages vs L")
plt.xlabel("L (# data per block)")
plt.ylabel("Error")
plt.xscale("log")
plt.plot(e_errors[:,0],e_errors[:,1])

#pressure errors
plt.subplot(1,2,2)
plt.title("Pressure - errors in averages vs L")
plt.xlabel("L (# data per block)")
plt.ylabel("Error")
plt.xscale("log")
plt.plot(p_errors[:,0],p_errors[:,1])

plt.show()
```

It is worth to be noticed that the error is monotonically increasing with L when $L \lesssim 50$. This means that the value of the uncertainty for small L is underestimated due to high correlation in the data which lower the variance.
For  $L \gtrsim 50$ the value of uncertainty decrease with L as expected. The critical value beteween the two behaviours is roughly twice the correlation time. For a large number of measures in a block the values of data are no more correlated and thus the incertainties decrease with the number of measures as predicted by the CLT. <br>
This is a further evidence that suggest us to use a big number of data in each block ($L \gtrsim 50$); the same result was found studying autocorrelation function above.
