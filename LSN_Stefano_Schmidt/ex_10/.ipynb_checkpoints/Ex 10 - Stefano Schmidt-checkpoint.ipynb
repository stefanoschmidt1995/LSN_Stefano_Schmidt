In **ex_10.cpp** it has been implemented the algorithm of Simulated Annealing. The algorithm makes use of a single istance of class `individual_type` (defined in **../GA_helper/GA_helpers.h** and described for exercise 9) to make a Metropolis algorithm where a proposed move is a random "genetic" mutation. The temperature is decreased with a power law: $T_n = T_0 \cdot (1+c)^{-n}$. The constant c is set by hand to a value of $c=0.05$ and $T_0=1000$. The algorithm stops when a number $N_{max}=100$ of temperature updates are done. From an update of temperature to another the algorithm performs $N_{step} = 1000$ Metropolis steps.<br>
The parallel code is written in **ex_10_parallel.cpp** and is developed in the same fashion as the noo parallel one.

### Exercise 10.1
The code in **ex_10.cpp** produces an output file **distances.dat** holding the best path lenght as a function of $T$ or as a function of the number of updates of temperature..<br>
The obtained best path is saved to file **best_path.dat** and plotted below.<br>
As for the previous exercise the plot of the best path lenght are done for both a configuration of 30 cities in a circle and in a box. The learning rate is computed for the box case.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#learning rate
distances = np.loadtxt("distances.dat", dtype=float)

plt.figure(figsize=(10,5))
plt.plot(distances[:,1], distances[:,2],label="best path lenght")
plt.xlabel("T",fontsize=10)
plt.xscale("log")
plt.ylabel("path lenght",fontsize=10)
plt.legend()
plt.show()
```

```python
#best path box
file = "best_path_box.dat"
best_path = np.loadtxt(file, dtype=float,skiprows=1)
with open(file) as f:
    best_path_lenght = f.readline()

plt.figure(figsize=(8,8))
plt.plot(best_path[:,0], best_path[:,1],'go--', label=best_path_lenght,  linewidth=2, markersize=10)
plt.annotate("0",(best_path[0,0],best_path[0,1]), xytext = (best_path[0,0]+.15,best_path[0,1]+.15), arrowprops=dict(facecolor='black', shrink=0.0001),fontsize=12)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#best path circle
file = "best_path_circle.dat"
best_path = np.loadtxt(file, dtype=float,skiprows=1)
with open(file) as f:
    best_path_lenght = f.readline()

plt.figure(figsize=(8,8))
plt.plot(best_path[:,0], best_path[:,1],'go--', label=best_path_lenght,  linewidth=2, markersize=10)
plt.annotate("0",(best_path[0,0],best_path[0,1]), xytext = (best_path[0,0]+.15,best_path[0,1]+.15), arrowprops=dict(facecolor='black', shrink=0.0001),fontsize=12)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

It is worth to be noticed that the solutions obtain with SA are better than those obtained by GA in terms of path lenght obtained (this happens both for a box and circle configuration). Furthermore the GA is much slower to run rathern than SA. Definetely SA works better in this case.

### Exercise 10.2
The code in **ex_10_parallel.cpp** works in the same way as the one in **ex_10.cpp**. It produces as output a file **best_path_parallel.dat** for both the configuration box and circle. The code has worked with $n=20$ cores.

```python
#best path box
file = "best_path_parallel_box.dat"
best_path = np.loadtxt(file, dtype=float,skiprows=1)
with open(file) as f:
    best_path_lenght = f.readline()

plt.figure(figsize=(8,8))
plt.plot(best_path[:,0], best_path[:,1],'go--', label=best_path_lenght,  linewidth=2, markersize=10, color="b")
plt.annotate("0",(best_path[0,0],best_path[0,1]), xytext = (best_path[0,0]+.15,best_path[0,1]+.15), arrowprops=dict(facecolor='black', shrink=0.0001),fontsize=12)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#best path circle
file = "best_path_parallel_circle.dat"
best_path = np.loadtxt(file, dtype=float,skiprows=1)
with open(file) as f:
    best_path_lenght = f.readline()

plt.figure(figsize=(8,8))
plt.plot(best_path[:,0], best_path[:,1],'go--', label=best_path_lenght,  linewidth=2, markersize=10, color="b")
plt.annotate("0",(best_path[0,0],best_path[0,1]), xytext = (best_path[0,0]+.15,best_path[0,1]+.15), arrowprops=dict(facecolor='black', shrink=0.0001),fontsize=12)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

The solution found by the parallel code is better than the one found by the ordinary code and probably is really close to the optimal one. This is no mistery: in the parallel code the algorithm has run 10 times while in the ordinary code only once; in 10 repetitions it is more probable to get a good solution to the TSP.
